Simular dos variables aleatorias $X_1$ y $X_2$ que siguen distribuciones de Poisson con parametros $\mu_1 = 1$ y $\mu_2 = 2$ respectivamente. Calcular la variable $Y = X_1 + X_2$. Repetir la simulaci칩n 1000 veces. Graficar la distribuci칩n de la variable $Y$ y comparar con una distribuci칩n de probabilidad adecuada.

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import poisson 
```

```python
# variables aleatorias con distribucion de Poisson:

mu1 = 1
mu2 = 2

X1 = poisson.rvs(mu1, size=1000)
X2 = poisson.rvs(mu2, size=1000)

# Calcular Y = X1 + X2:

Y = X1 + X2
```

```python
# Graficar la distribuci칩n de Y:

bins = np.arange(min(Y)-0.5, max(Y) + 1.5)

plt.figure(figsize=(10, 6))
plt.hist(Y, bins = bins, alpha=0.6, edgecolor='black', linewidth=1.2, density=True, label='Simulaci칩n Y')

# Comparar con una distribuci칩n de Poisson con par치metro mu = mu1 + mu2:

mu_Y = mu1 + mu2
x = np.arange(0, max(Y) + 1)
poisson_pmf = poisson.pmf(x, mu_Y)
plt.plot(x, poisson_pmf, 'ro-', label=f'Poisson PMF (mu={mu_Y})')

plt.xlabel('Y')
plt.ylabel('Densidad')
plt.title('Y = X1 + X2')
plt.legend()
plt.show()
```
